## Print contents of csv to PDF

```python
import win32com.client

def excel_to_pdf(wb_path,path_to_pdf):

    #o = win32com.client.Dispatch("Excel.Application")
    
    o = win32com.client.gencache.EnsureDispatch("Excel.Application")

    o.Visible = False

    #parameter
    #wb_path = r'C:\Users\M Aldeza\Documents\Arthur Lagura\2_New Clark City\AsBuilt\YEAR 2022\test_folder\test_As-Alfie 2A BL SW 220722.csv'

    wb = o.Workbooks.Open(wb_path)

    ws_index_list = [1] #say you want to print these sheets

    #parameter
    #path_to_pdf = r'C:\Users\M Aldeza\Documents\Arthur Lagura\2_New Clark City\AsBuilt\YEAR 2022\test_folder\test_As-Alfie 2A BL SW 220722.pdf'

    print_area = 'A:F'

    test=Path(wb_path)
    file_name=test.stem


    for index in ws_index_list:

        #off-by-one so the user can start numbering the worksheets at 1

        #ws = wb.Worksheets[index]
        ws=wb.Sheets(index)
        ws.Columns("B:C").NumberFormat = '0.000'
        
        #for inserting label
        const = win32com.client.constants
        
        ws.Range("A1:F1").Insert(Shift=const.xlShiftDown)
        
        rangeObj=ws.Range("A1:A1")
        rangeObj.Value=f'{file_name}'
        rangeObj.EntireRow.Insert(Shift=const.xlShiftDown)
        
        
        #for print setup
#         ws.PageSetup.Zoom = False

#         ws.PageSetup.FitToPagesTall = 2

#         ws.PageSetup.FitToPagesWide = 1

#         ws.PageSetup.PrintArea = print_area


    wb.Sheets(ws_index_list).Select()

    wb.ActiveSheet.ExportAsFixedFormat(0, path_to_pdf)
    
    wb.Close(True)

    
    
```
